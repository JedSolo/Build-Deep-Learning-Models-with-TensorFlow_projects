### Importing all the necessary libraries

```python
import pandas as pd
import numpy as np
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, InputLayer
from sklearn.metrics import classification_report
```

### Loading the data

```python
data = pd.read_csv('heart_failure.csv')
data.head(10)
```

```python
data.info()
```

```python
# Checking the distribution of the death_event column:
print('Classes and number of values in the dataset',Counter(data['death_event']))
```

### Getting the labels and features

```python
# Assigning the label:
y = data['death_event']

# Assigning the features:
x = data[['age','anaemia','creatinine_phosphokinase','diabetes','ejection_fraction','high_blood_pressure','platelets','serum_creatinine','serum_sodium','sex','smoking','time']]
```

### Data Preprocessing

```python
# Converting all the categorical columns to one-hot encoding vectors:
x = pd.get_dummies(x)

# Splitting the data into train set and test set:
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)

# Scaling the numeric features:
ct = ColumnTransformer([("numeric", StandardScaler(), ['age','creatinine_phosphokinase','ejection_fraction','platelets','serum_creatinine','serum_sodium','time'])])
X_train = ct.fit_transform(X_train)
X_test = ct.transform(X_test)
```

### Prepare labels for classification

```python
# Lable encoding the label column:
le = LabelEncoder()
Y_train = le.fit_transform(Y_train.astype(str))
Y_test = le.transform(Y_test.astype(str))

Y_train = to_categorical(Y_train)
Y_test = to_categorical(Y_test)
```

### Design the Model

```python
model = Sequential()

# Adding input layer:
model.add(InputLayer(input_shape=(X_train.shape[1],)))

# Adding a hidden layer:
model.add(Dense(12, activation='relu'))

# Adding an output layer:
model.add(Dense(2, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

### Train and evaluate the model

```python
model.fit(X_train, Y_train, epochs = 100, batch_size = 16, verbose=1)
```

```python
# Evaluating the model:
loss, acc = model.evaluate(X_test, Y_test, verbose=0)
print("Loss", loss, "Accuracy:", acc)
```

### Generating a classification report

```python
y_estimate = model.predict(X_test, verbose=0)
y_estimate = np.argmax(y_estimate, axis=1)
y_true = np.argmax(Y_test, axis=1)
print(classification_report(y_true, y_estimate))
```
